```python
from __future__ import print_function
import sys
import linecache
import numpy as np
import time
import math 
# import tensorflow as tf
import tensorflow.compat.v1 as tf 
tf.disable_v2_behavior()

train_file = "/Volumes/TOSHIBA/dbpedia/0414/data/dbpedia.train"
test_file = "/Volumes/TOSHIBA/dbpedia/0414/data/dbpedia.test"
label_dict = {}
sku_dict = {}
```

```python
max_window_size = 1000
batch_size = 500
emb_size = 128

# Parameters
learning_rate = 0.01
training_epochs = 1
display_step = 1

# Network Parameters
n_hidden_1 = 128 # 1st layer number of features
# n_hidden_2 = 256 # 2nd layer number of features
```

```python
def init_data(read_file):
    """
    对每一行Word 打标签(播放时长)，并且对 word 汇总编码 
    label_dict: 14类, 全部的播放时长标签
    sku_dict:  803107个word集合，并编码(按序号)  
    """
    #0 is used for padding embedding
    label_cnt = 0
    sku_cnt = 1
    f = open(read_file,'r')
    for line in f:
        line = line.strip().split(' ')
        for i in line:
            if i.find('__label__') == 0:   #每行做标签，按顺序依次
                if i not in label_dict:
                    label_dict[i] = label_cnt
                    label_cnt += 1
            else:
                if i not in sku_dict:
                    sku_dict[i] = sku_cnt
                    sku_cnt += 1
```

```python
def read_data(pos, batch_size, data_lst):
    """
    batch_size: 500行是一批batch 
    pos : 第几批次
    max_window_size: 每行最大的word 个数 
    
    x:  每一批次(每行word序列,已经编码)      （batch_size, max_window_size)
    y:  每一批次(每行标签)      np.array(y) (batch_size, 1)
    mask: 每一批次(每行word序列, 编码全为1)  （batch_size, max_window_size)
    word_num: 每一批次(每行序列长度)         (batch_size, 1)
    
    """
    batch = data_lst[pos:pos + batch_size]
    x = np.zeros((batch_size, max_window_size))
    mask = np.zeros((batch_size, max_window_size))
    y = []
    word_num = np.zeros((batch_size))
    line_no = 0
    for line in batch:
        line = line.strip().split(' ')
        y.append(label_dict[line[0]])
        col_no = 0
        for i in line[1:]:
            if i in sku_dict:
                x[line_no][col_no] = sku_dict[i]  # 每一行Word序列,按Word 顺序编码
                mask[line_no][col_no] = 1         # 每一行Word序列,取1作为排序编码, 最大长度不超过 1000维
                col_no += 1                       # 序列加1
            if col_no >= max_window_size:         # 单行序列, 最多word 个数，不超过1000列    
                break
        word_num[line_no] = col_no                # 每一行Word 序列, 序列长度, word可重复 
        line_no += 1

    return  x, \
            np.array(y).reshape(batch_size, 1), \
            mask.reshape(batch_size, max_window_size,1), \  
            word_num.reshape(batch_size, 1)
```

```python
# Store layers weight & bias
weights = {
    'h1': tf.Variable(tf.random_normal([emb_size, n_hidden_1])),
    # 'h2': tf.Variable(tf.random_normal([n_hidden_1, n_hidden_2])),
    'out': tf.Variable(tf.random_normal([n_hidden_1, n_classes]))
}
biases = {
    'b1': tf.Variable(tf.random_normal([n_hidden_1])), 
    # 'b2': tf.Variable(tf.random_normal([n_hidden_2])),
    'out': tf.Variable(tf.random_normal([n_classes])) 
}
```

```python
# Create model
def multilayer_perceptron(x, weights, biases):
    # Hidden layer with RELU activation
    # x = tf.nn.dropout(x, 0.8)
    layer_1 = tf.add(tf.matmul(x, weights['h1']), biases['b1'])
    layer_1 = tf.nn.relu(layer_1)
    
    # dlayer_1 = tf.nn.dropout(layer_1, 0.5)
    # layer_2 = tf.add(tf.matmul(layer_1, weights['h2']), biases['b2'])
    # layer_2 = tf.nn.relu(layer_2)
    
    # Output layer with linear activation
    # out_layer = tf.matmul(layer_1, weights['out']) + biases['out']
    # return out_layer
    
    return layer_1
```

```python
embedding = {
    'input':tf.Variable(tf.random_uniform([len(sku_dict)+1, emb_size], -1.0, 1.0))
    # 'output':tf.Variable(tf.random_uniform([len(label_dict)+1, emb_size], -1.0, 1.0))
}
```

```python
emb_mask = tf.placeholder(tf.float32, shape=[None, max_window_size, 1])
word_num = tf.placeholder(tf.float32, shape=[None, 1])

x_batch = tf.placeholder(tf.int32, shape=[None, max_window_size])
y_batch = tf.placeholder(tf.int64, [None, 1])
```

```python
input_embedding = tf.nn.embedding_lookup(embedding['input'], 
                                         x_batch)
```

```python
project_embedding = tf.div(tf.reduce_sum(tf.multiply(input_embedding,emb_mask), 1),
                           word_num)
```

```python
# Construct model
pred = multilayer_perceptron(project_embedding, 
                             weights, 
                             biases)
```

```python
# Construct the variables for the NCE loss
nce_weights = tf.Variable(tf.truncated_normal([n_classes, n_hidden_1], 
                                               stddev=1.0/math.sqrt(n_hidden_1)))
nce_biases = tf.Variable(tf.zeros([n_classes]))
loss = tf.reduce_mean(tf.nn.nce_loss(weights=nce_weights,
                                     biases=nce_biases,
                                     labels=y_batch,
                                     inputs=pred,
                                     num_sampled=10,
                                     num_classes=n_classes))
cost = tf.reduce_sum(loss) /batch_size
optimizer = tf.train.AdamOptimizer(learning_rate=learning_rate).minimize(cost)
out_layer = tf.matmul(pred, tf.transpose(nce_weights)) + nce_biases
```

```python
init = tf.global_variables_initializer()
```

```python
#with tf.Session(config=tf.ConfigProto(log_device_placement=True)) as sess:
with tf.Session() as sess:
    sess.run(init)

    # Training cycle
    start_time = time.time()
    total_batch = int(len(train_lst) / batch_size)
    print("total_batch of training data: ", total_batch)
    for epoch in range(training_epochs):
        avg_cost = 0.
        for i in range(total_batch):
            x, y, batch_mask, word_number = read_data(i * batch_size, batch_size, train_lst)
            _,c = sess.run([optimizer, cost], feed_dict={x_batch: x, 
                                                         emb_mask: batch_mask, 
                                                         word_num: word_number, 
                                                         y_batch: y})
            print("Epoch %d Batch %d Elapsed time %fs" %(epoch, i, time.time() - start_time))
            # Compute average loss
            avg_cost += c / total_batch
            correct_prediction = tf.equal(tf.argmax(out_layer, 1), 
                                          tf.reshape(y_batch, [batch_size]))
            accuracy = tf.reduce_mean(tf.cast(correct_prediction, "float"))
            print("Accuracy:", accuracy.eval({x_batch: x, 
                                              y_batch: y, 
                                              emb_mask: batch_mask, 
                                              word_num: word_number}))

        # Display logs per epoch step
        if epoch % display_step == 0:
            print("Epoch:", '%04d' % (epoch + 1), "cost=", "{:.9f}".format(avg_cost))

    # Test model
    correct_prediction = tf.equal(tf.argmax(out_layer, 1), 
                                  tf.reshape(y_batch, [batch_size]))
    # Calculate accuracy
    accuracy = tf.reduce_mean(tf.cast(correct_prediction, "float"))

    test_lst = linecache.getlines(test_file)
    total_batch = int(len(test_lst) / batch_size)
    final_accuracy = 0
    for i in range(total_batch):
        x, y, batch_mask, word_number = read_data(i*batch_size, batch_size, test_lst)
        batch_accuracy = accuracy.eval({x_batch: x, y_batch: y, emb_mask: batch_mask, word_num: word_number})
        print("Batch Accuracy: ", batch_accuracy)
        final_accuracy += batch_accuracy
    print("Final Accuracy: ", final_accuracy * 1.0 / total_batch)
```
